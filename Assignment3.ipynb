# **problem 1**

```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
high_score = max(employees.values())
top_employees = []
for name, score in employees.items():
    if score == high_score:
        top_employees.append(name)
print("Top Performers Eligible for Bonus:", ", ".join(top_employees), f"(Score: {high_score})")
```

# **problem 2**

```python
S = "Buy mobile phone buy phone online"
query = S.lower()

#split into words
words = query.split()

#count words
word_count = {}
for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

# keep only words with count > 1
result = {}
for word, count in word_count.items():
    if count > 1:
        result[word] = count

print(result)
```

# **problem  3**

```python
readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for hour, value in enumerate(readings):
    if value % 2 == 0:   # check if even
        valid_readings.append((hour, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

# **problem 4**

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
domains = [email.split("@")[1] for email in emails]
domain_counts = {}
for d in domains:
    domain_counts[d] = domain_counts.get(d, 0) + 1
total = len(emails)
for domain, count in domain_counts.items():
    percentage = (count / total) * 100
    print(f"{domain}: {percentage:.0f}%")
```

# **problem 5**

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

average = sum(sales) / len(sales)
threshold = 1.3 * average

print("Average sales:", average)
print("Threshold (30% above average):", threshold)

for day, value in enumerate(sales, start=1):
    if value >= round(threshold, -2):
        print(f"Day {day}: {value}")
```

explaination why we cant get day4 output at first:- Average sales = 10200/6=1700
- Threshold (30% above average) = 1.3\times 1700=2210
Now check Day 4:
- Day 4 sales = 2200
- Threshold = 2210
- Since 2200 is less than 2210, it doesn’t qualify.
That’s why Python doesn’t print Day 4.

```python
def valid_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print(" Recharge amount must be at least ₹50.")
        return False
    elif amount not in valid_plans:
        print("Invalid plan. Valid plans are:", valid_plans)
        return False
    else:
        print(" Recharge successful for ₹", amount)
        return True


# program with retry option
while True:
    try:
        amount = int(input("Enter recharge amount: ₹"))
        if valid_recharge(amount):
            break  # Exit loop if recharge is valid
        else:
            print(" tryagain.\n")
    except ValueError:
        print(" Invalid input. enter a number.\n")
```

```python
def check_inventory(stock_dict):
    for product, quantity in stock_dict.items():
        if quantity < 15:
            print(f"{product}: Reorder Alert (Stock = {quantity})")
        else:
            print(f"{product}: Stock OK (Stock = {quantity})")


# Example dictionary of products and stock quantities
inventory = {
    "toothpaste": 10,
    "lipbalm": 25,
    "talc": 14,
    "makeupkit": 30,
    "shoes": 5
}
check_inventory(inventory)
```

```python
def process_result(marks_list):
    total = 0
    for mark in marks_list:
        total += mark

    avg= total / len(marks_list)
    print("Average Marks:", avg)

    if avg >= 50:
        return "Pass"
    else:
        return "Fail"
student_marks = [60, 45, 70, 55, 40]  # sample marks
result = process_result(student_marks)
print("Result:", result)
```

```python
def attendance_check(attend_list):
    # Count present days
    present_days = 0
    for day in attend_list:
        if day == "P":
            present_days += 1

    # Calculate percentage
    total_days = len(attend_list)
    attend_percentage = (present_days / total_days) * 100

    # Return eligibility
    if attend_percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"
attendance = ["P", "A", "P", "P", "A", "P", "P", "P"]
print(attendance_check(attendance))
```

```python
def check_password(password):
    if len(password) < 8:
        return "Too short! Make it at least 8 characters."

    if not any(ch.isdigit() for ch in password):
        return "Add at least one number."

    if not any(ch in "@#$" for ch in password):
        return "Add at least one special symbol (@, #, $)."

    return "Nice! Your password looks strong."
print(check_password("hello"))
print(check_password("helloworld"))
print(check_password("hello1234"))
print(check_password("hello1234@"))
```

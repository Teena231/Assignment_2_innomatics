```python
def smart_park(capacity,logs):
  parked=0
  peak=-0
  for log in logs:
    if log=="IN":
      parked+=1
    elif log=="OUT" and parked> 0:
      parked-=1
    peak=max(peak,parked)
  status="Full" if parked>=capacity else "Available"
  print("Currently parked vehicles:",parked)
  print("Parking Status",status)
smart_park(50,["IN","IN","IN","OUT","IN","IN","OUT"])
```

```python
def delivery(distance,traffic,weather):
  time =distance * 5
  if traffic=="High":
    time+=15
  elif traffic=="Medium":
    time+=10
  if weather=="Stormy":
    time+=20
  print("Estimated Delivery Time:",time,"Minutes")
delivery(8,"High","Rainy")
```

```python
def theatre(total_seats, booked):
    occupancy = (len(booked) / total_seats) * 100
    if occupancy == 100:
        stat= "Housefull"
    elif occupancy >= 75:
        stat = "Almost Full"
    else:
        stat = "Available"
    print("Occupancy:", f"{occupancy:.0f}%")
    print("Show Status:", stat)
theatre(200, [1]*150)
```

```python
def server_load(readings):
    avg = sum(readings) / len(readings)
    if avg < 50: status = "Normal"
    elif avg <= 80: status = "Warning"
    else: status = "Critical"
    print("Average CPU Load:", f"{avg:.0f}%")
    print("Server Status:", status)
server_load([45,60,70,85,90])
```

```python
def resource_monitor(usage):
    overused = [r for r,h in usage.items() if h > 8]
    alert = "Yes" if overused else "No"
    print("Overused Resources:", ", ".join(overused) if overused else "None")
    print("Energy Alert:", alert)
resource_monitor({"Projector":6,"AC":9,"Lights":4})
```

```python
def event_registration(capacity, registrations):
    confirmed = min(capacity, registrations)
    waitlist = max(0, registrations - capacity)
    status = "Closed" if registrations >= capacity else "Open"
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)
event_registration(100, 105)
```
